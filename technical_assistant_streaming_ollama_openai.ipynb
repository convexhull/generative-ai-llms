```python
# imports
from openai import OpenAI
from dotenv import load_dotenv
import os
from IPython.display import Markdown, display, update_display
```

```python
# constants

MODEL_GPT = 'gpt-4o-mini'
MODEL_LLAMA = 'llama3.2'
```

```python
# set up environment
load_dotenv(override=True)
print(os.getenv("OPENAI_API_KEY"))
```

```python
# here is the question; type over this to ask something new

question = """
Please explain what this code does and why:
yield from {book.get("author") for book in books if book.get("author")}
"""
```

```python
# Get Llama 3.2 to answer# Get gpt-4o-mini to answer, with streaming
openai = OpenAI()

# openai = OpenAI(base_url="http://localhost:11434/v1", api_key="")

system_prompt = """
You are a technical assistant. Your job is to answer technical questions to a beginner in the field. You rich markdown for response
"""

user_prompt = f"""
Here is first question: {question}
"""

print(user_prompt)

messages = [
    {"role" : "system" , "content" : system_prompt},
    {"role": "user", "content" : user_prompt}
]

stream = openai.chat.completions.create(model=MODEL_GPT, messages=messages, stream=True)

# for chunk in stream:
#     delta = chunk.choices[0].delta.content
#     if delta:
#         response += delta
#         print(delta, end="")


response = ""
display_handle = display(Markdown(""), display_id=True)
for chunk in stream:
        response += chunk.choices[0].delta.content or ''
        update_display(Markdown(response), display_id=display_handle.display_id)

```
