```python
openai = OpenAI()
```

```python
def shout(text):
    return text.upper()
```

```python
gr.Interface(fn=shout, inputs="textbox", outputs="textbox").launch(inbrowser=True, auth=[('yash','pwd1'), ('harsh','pwd2')])
```

```python
system_prompt = """You are a question answer agent. You have to respond to user's question by treating them as beginners in the concerned field. 
Respond in markdown without codeblocks.
"""

```

```python
def message_gpt(message):
    user_prompt = message
    response = openai.chat.completions.create(model="gpt-4.1-mini", messages=[{"role": "system", "content": system_prompt},{"role": "user", "content": user_prompt}])
    return response.choices[0].message.content
    
```

```python
def stream_gpt(message):
    user_prompt = message
    response = ""
    stream = openai.chat.completions.create(model="gpt-4.1-mini", messages=[{"role": "system", "content": system_prompt},{"role": "user", "content": user_prompt}], stream=True)
    for chunk in stream:
        response += chunk.choices[0].delta.content or ""
        yield response
```

```python
message_gpt("howdie!")
```

```python
chat_input = gr.Textbox(label="Your message", info="Enter a message!")
chat_output = gr.Markdown(label="Response:")

interface = gr.Interface(inputs=[chat_input], outputs=[chat_output], fn = message_gpt)

interface.launch(inbrowser=True)
```

```python
chat_input = gr.Textbox(label="Your message", info="Enter a message!")
chat_output = gr.Markdown(label="Response:")

interface = gr.Interface(inputs=[chat_input], outputs=[chat_output], fn = stream_gpt)

interface.launch(inbrowser=True)
```
