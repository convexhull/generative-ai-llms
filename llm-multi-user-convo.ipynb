```python
alex_system_prompt = """
You are Alex, a chatbot who is very argumentative; you disagree with anything in the conversation and you challenge everything, in a snarky way.
You are in a conversation with Blake and Charlie. Just say your reply, don't put any convo annotation for who is speaking
"""

blake_system_prompt = """
You are Blake, a chatbot who is very excited; you are impressed with anything in the conversation and you hype everything, in a active way.
You are in a conversation with Blake and Charlie. Just say your reply, don't put any convo annotation for who is speaking
"""

charlie_system_prompt = """
You are Charlie, a chatbot who is very pessimistic; you are pessimistic with anything in the conversation and you calm down everything, in a cynical way.
You are in a conversation with Blake and Charlie. Just say your reply, don't put any convo annotation for who is speaking
"""

convo = """
Alex: Hi
Blake: Hello!
Charlie: Hey\n"""

def chat_alex():
    global convo
    alex_user_prompt = f"""
    You are Alex, in conversation with Blake and Charlie.
    The conversation so far is as follows:
    {convo}
    Now with this, respond with what you would like to say next, as Alex.
    """
    messages = [
        {"role": "system", "content": alex_system_prompt},
        {"role": "user", "content" : alex_user_prompt}
    ]
    response = openai.chat.completions.create(model="gpt-4.1-mini", messages=messages)
    chirp = f"Alex: {response.choices[0].message.content}\n"
    convo += f"""{chirp}"""
    print(chirp)

def chat_blake():
    global convo
    blake_user_prompt = f"""
    You are Blake, in conversation with Alex and Charlie.
    The conversation so far is as follows:
    {convo}
    Now with this, respond with what you would like to say next, as Blake.
    """
    messages = [{
        "role": "system", "content": blake_system_prompt},
        {"role": "user", "content" : blake_user_prompt
    }]
    response = openai.chat.completions.create(model="gpt-4.1-mini", messages=messages)
    chirp = f"Blake: {response.choices[0].message.content}\n"
    convo += f"""{chirp}"""
    print(chirp)

def chat_charlie():
    global convo
    charlie_user_prompt = f"""
    You are Charlie, in conversation with Alex and Blake.
    The conversation so far is as follows:
    {convo}
    Now with this, respond with what you would like to say next, as Charlie.
    """
    messages = [{
        "role": "system", "content": charlie_system_prompt},
        {"role": "user", "content" : charlie_user_prompt
    }]
    response = openai.chat.completions.create(model="gpt-4.1-mini", messages=messages)
    chirp = f"Charlie: {response.choices[0].message.content}\n"
    convo += f"""{chirp}"""
    print(chirp)

for i in range(5):
    chat_alex()
    chat_blake()
    chat_charlie()
```
