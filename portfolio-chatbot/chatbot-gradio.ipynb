```python
import os
from dotenv import load_dotenv
from openai import OpenAI
from scraper import fetch_website_contents
import gradio as gr
```

```python
load_dotenv(override=True)
```

```python
print(os.getenv("OPENAI_API_KEY"))
```

```python
openai = OpenAI()
```

```python
system_prompt = """
You are a chatbot that answers questions. Your main domain is to answer questions about developer's 
portfolios. You will be provided scraped info from developer's portfolio website. Keep your answers entertaining"""
```

```python
portfolio_data = fetch_website_contents("https://yashpratapsingh.com")
```

```python
def stream_portfolio_chat(message, history):
    updated_system_prompt = system_prompt + f"""\nHere is the developer portfolio scraped info:
    {portfolio_data}
    """
    sanitized_history = [{"role": h["role"], "content": h["content"]} for h in history]
    messages = [{"role" : "system", "content" : updated_system_prompt}] + sanitized_history + [{"role": "user", "content" : message}]
    stream = openai.chat.completions.create(model="gpt-4.1-mini", messages=messages, stream=True)
    response = ""
    for chunk in stream:
        response += chunk.choices[0].delta.content or ""
        yield response
```

```python
gr.ChatInterface(type="messages", fn=stream_portfolio_chat).launch(inbrowser=True)
```
