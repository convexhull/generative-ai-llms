```python
import os
from openai import OpenAI
from dotenv import load_dotenv
import gradio as gr
```

```python
load_dotenv(override=True)
print(os.getenv("OPENAI_API_KEY"))
```

```python
openai = OpenAI()
```

```python
system_prompt = """
You are a agent that is supposed to answer user's questions in a meaningful but also sarcastic way
"""
```

```python
def chat_gpt(message, history):
    messages = [{"role" : "system", "content": system_prompt}] + history + [{"role": "user", "content": message}]
    response = openai.chat.completions.create(model="gpt-4.1-mini", messages=messages)
    return response.choices[0].message.content
```

```python
def stream_gpt(message, history):
    # sanitization for non-openai llms if needed
    history = [{"role": h["role"], "content": h["content"]} for h in history]
    messages = [{"role" : "system", "content": system_prompt}] + history + [{"role": "user", "content": message}]
    stream = openai.chat.completions.create(model="gpt-4.1-mini", messages=messages, stream=True)
    response = ""
    for chunk in stream:
        response += chunk.choices[0].delta.content or ""
        yield response
```

```python
gr.ChatInterface(fn=stream_gpt, type="messages", flagging_mode="never").launch(inbrowser=True)
```

```python
gr.ChatInterface(fn=chat_gpt, type="messages", flagging_mode="never").launch()
```
